```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('dataset.csv')
data
```

```python
data = data.dropna(subset=["ship.name"])
```

```python
data[data["ship.name"] == "Жужа"]
```

```python
shipname = data["ship.name"].unique()
data[data["ship.name"] == "Жужа"][["dock.name", "Причаливание", "Отход"]].values
```

```python
def seconder(s:str):
    mas = s.split()
    t = mas[0].split(':')
    hours = int(t[0])
    minutes = int(t[1])
    seconds = int(t[2])
    if mas[1] == 'PM':
        hours+=12
    sec = seconds + minutes*60 + hours*3600
    return sec
```

```python
mas = data[data["ship.name"] == "Жужа"][["dock.name", "Причаливание", "Отход"]].values
```

```python
for i in range(len(mas)):
    mas[i][1] = seconder(mas[i][1])
    mas[i][2] = seconder(mas[i][2])
mas
```

```python
sl = dict()
for i in range(1, len(mas)):
    k = (mas[i-1][0], mas[i][0])
    actiontime = mas[i][1] - mas[i-1][2]
    if k in sl:
        sl[k].append(actiontime)
    else:
        sl[k] = [actiontime]
```

```python
sl = dict()
```

```python
for n in shipname:
    nmas = data[data["ship.name"] == n].dropna(subset=["ship.name"])[["dock.name", "Причаливание", "Отход"]].values
    if type(n) != str:
        continue
    for i in range(len(nmas)):
        if type(nmas[i, 1]) != str:
            continue
        nmas[i][1] = seconder(nmas[i][1])
        nmas[i][2] = seconder(nmas[i][2])
    for i in range(1, len(nmas)):
        k = (nmas[i-1][0], nmas[i][0])
        if type(nmas[i, 1]) != int or type(nmas[i - 1, 2]) != int:
            continue
        actiontime = nmas[i][1] - nmas[i-1][2]
        if actiontime <= 0:
             continue
        if k in sl:
            sl[k].append(actiontime)
        else:
            sl[k] = [actiontime]
nmas
```

```python
sl
```

```python
for i in sl.keys():
      sl[i] = np.mean(sl[i])
sl
```

```python
dockname = data["dock.name"].unique()
dockname
```

```python
d = dict()
for i in dockname:
    d[i] = dict()
    for j in dockname:
        d[i][j] = np.inf
        if i == j:
            d[i][j] = 0
        elif (i, j) in sl:
            d[i][j] = sl[(i,j)]

for k in dockname:
	for i in dockname:
		for j in dockname:
			d[i][j] = min (d[i][j], d[i][k] + d[k][j]);
d
```

```python
import matplotlib.pyplot as plt
```

```python
# Преобразуем словарь в массив
rows = list(d.keys())
cols = list(next(iter(d.values())).keys())

def process_value(value):
    rounded = np.round(value / 60)  
    return rounded if rounded <= 1000 else np.nan  

data = np.array([[process_value(d[row][col]) for col in cols] for row in rows])


fig, ax = plt.subplots(figsize=(len(cols) * 0.5, len(rows) * 0.5))  # Увеличиваем ширину фигуры
cax = ax.matshow(data, cmap='coolwarm', vmin=0, vmax=200)


cbar = fig.colorbar(cax)
cbar.set_label("Время (минуты)", fontsize=12)


ax.set_xticks(range(len(cols)))
ax.set_yticks(range(len(rows)))
ax.set_xticklabels(cols, rotation=90, ha='center', fontsize=8)  # Угол 90° и уменьшенный шрифт
ax.set_yticklabels(rows, fontsize=8)
ax.set_xlabel("Откуда", fontsize=12, labelpad=10)
ax.set_ylabel("Куда", fontsize=12, labelpad=10)


for i in range(len(rows)):
    for j in range(len(cols)):
        value = data[i, j]
        if not np.isnan(value):  # Отображать только числа, которые не NaN
            ax.text(j, i, f'{int(value)}', ha='center', va='center', color='black', fontsize=6)


plt.title("Время в минутах, чтобы добраться от одного причала к другому", fontsize=14, pad=20)


plt.subplots_adjust(left=0.2, right=0.9, top=0.85, bottom=0.35)


output_file = "table_visualization_adjusted.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')
plt.show()

print(f"Таблица сохранена в файл: {output_file}")
```
